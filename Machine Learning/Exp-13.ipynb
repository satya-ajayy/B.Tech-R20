### Experiment-13: 
Write a Python program to construct a Bayesian network considering medical 
data. Use this model to demonstrate the diagnosis of heart patients using 
standard Heart Disease Data Set

```python
import numpy as np
import pandas as pd
import csv 
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.models import BayesianNetwork
from pgmpy.inference import VariableElimination
```

```python
heartDisease = pd.read_csv('../All Data_Sets/heart_disease.csv')
heartDisease = heartDisease.replace('?',np.nan)
heartDisease.head()
```

```python
model= BayesianNetwork(
                        [('age','target'),
                        ('sex','target'),
                        ('exang','target'),
                        ('cp','target'),
                        ('target','restecg'),
                        ('target','chol')])

print('Learning CPD using Maximum likelihood estimators')
model.fit(heartDisease,estimator=MaximumLikelihoodEstimator)

print('Inferencing with Bayesian Network:')
HeartDiseasetest_infer = VariableElimination(model)

print('1.Probability of HeartDisease given evidence = restecg')
q1=HeartDiseasetest_infer.query(variables=['target'],evidence={'restecg':1})
print(q1)

print('2.Probability of HeartDisease given evidence = cp ')
q2=HeartDiseasetest_infer.query(variables=['target'],evidence={'cp':2})
print(q2)
```
