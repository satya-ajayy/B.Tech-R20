### Imports

```python
import spacy
from spacy import displacy
import tabulate as tb
nlp = spacy.load('en_core_web_sm')
```

```python
interest_text = 'He is interested in learning Natural Language Processing.'
interest_doc = nlp(interest_text)
displacy.render(interest_doc, style='dep')
```

```python
with open ("../All DataFiles/wiki_mlk.txt", "r") as f:
    text = f.read()
# print(text)
```

```python
doc = nlp(text)
doc
```

```python
lst=[[ent.text, ent.label_] for ent in doc.ents]
header=['Text', 'Label']
print(tb.tabulate(lst, headers=header,tablefmt='pretty'))
```
